```python
def string2tstamp(vector,formatString,startZero=True):
    timestamp =  np.array(map(lambda x: time.mktime(datetime.strptime(x,formatString).timetuple()),vector)).flatten()
    if startZero:
        timestamp = timestamp - np.min(timestamp)
    return timestamp

def normalize(vector):
    vNorm = (vector - np.mean(vector))/np.std(vector)
    #vNorm = (vector)/np.std(vector)
    return vNorm

def smoothing(vector,smooth=10):
    smoothV = np.convolve(vector,np.zeros([smooth])+1./smooth)[:-smooth+1]
    return smoothV
```
